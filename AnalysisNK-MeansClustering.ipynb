```python
# Importing the libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading the math performance file and storing the data.
df_mat = pd.read_csv('student-mat.csv')
df_mat.shape
```

```python
# Reading the portugese performance file and storing the data.
df_por = pd.read_csv('student-por.csv')
df_por.shape
```

```python
# Concatenating both data and storing in the df.
df= pd.concat([df_mat, df_por], ignore_index= True)
df.shape
```

```python
# Checking the first five rows of the df.
df.head()
```

```python
# Checking the last five rows of the df.
df.tail()
```

```python
# Checking the summary of the df.
df.info()
```

```python
# Checking the summary statistics of the df.
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initializing the LabelEncoder
label_encoder = LabelEncoder()

# Encoding the categorical column and storing in the same column
df['school'] = label_encoder.fit_transform(df['school'])
df['sex'] = label_encoder.fit_transform(df['sex'])
df['address'] = label_encoder.fit_transform(df['address'])
df['famsize'] = label_encoder.fit_transform(df['famsize'])
df['Pstatus'] = label_encoder.fit_transform(df['Pstatus'])
df['Medu'] = label_encoder.fit_transform(df['Medu'])
df['Fedu'] = label_encoder.fit_transform(df['Fedu'])
df['Mjob'] = label_encoder.fit_transform(df['Mjob'])
df['Fjob'] = label_encoder.fit_transform(df['Fjob'])
df['reason'] = label_encoder.fit_transform(df['reason'])
df['guardian'] = label_encoder.fit_transform(df['guardian'])
df['schoolsup'] = label_encoder.fit_transform(df['schoolsup'])
df['famsup'] = label_encoder.fit_transform(df['famsup'])
df['paid'] = label_encoder.fit_transform(df['paid'])
df['activities'] = label_encoder.fit_transform(df['activities'])
df['nursery'] = label_encoder.fit_transform(df['nursery'])
df['higher'] = label_encoder.fit_transform(df['higher'])
df['internet'] = label_encoder.fit_transform(df['internet'])
df['romantic'] = label_encoder.fit_transform(df['romantic'])
df
```

```python
# Checking the summary statistics of the df.
df.describe()
```

```python
# Plotting a scatterplot using seaborn.
sns.set_theme()
scaplot = sns.scatterplot(data = df, x = 'absences', y = 'G3')
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Selecting the features to perform clustering
X = df[['absences', 'G3']]

# Initializing KMeans with random_state 42
km = KMeans(random_state=42)

# Initializing the KElbowVisualizer for finding out the number of cluster.
visualizer = KElbowVisualizer(km, k=(2,10))

# Fit the data to the visualizer
visualizer.fit(X)

# Finalize and render the figure
visualizer.show()        
```

```python
# selecting the columns for k-means clustering
selected_columns = ['G3', 'absences']
X = df[selected_columns]

# Determines the number of clusters (k)
k = 5 

# Applying K-means clustering
km = KMeans(n_clusters=k, random_state=0)
km.fit(X)

# Getting the cluster labels for each data point
labels = km.labels_

# Adding the cluster labels to the original dataset
df['cluster'] = labels

df[['G3', 'absences', 'cluster']].head()
```

```python
# checking the unique clusters.
df.cluster.unique()
```

```python
# checking the data distribution per cluster.
pd.Series(km.labels_).value_counts().sort_index()
```

```python
# Selecting the columns for visualization
viz_columns = ['G3', 'absences']

# Plotting the boxplots for each cluster
for column in viz_columns:
    plt.figure(figsize=(8, 8))
    sns.boxplot(x = 'cluster', y=column, data=df)
    plt.title(f'{column} by cluster.')
    plt.show()
```

```python
# Selecting the columns to visualize the clusters.
x_column = 'absences'
y_column = 'G3'

# Creating a scatter plot using Seaborn.
plt.figure(figsize=(8, 8))
scatterplot = sns.scatterplot(data=df, x=x_column, y=y_column, hue='cluster', palette='viridis')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Clusters Visualization')
plt.show()
```

```python
# Grouping and checking the median of absences and G3 per cluster.
df.groupby('cluster').agg({
    'absences':'median',
    'G3':'median'
})
```
